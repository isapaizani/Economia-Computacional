```python
"""
Programa adulto
Descrição: Este programa pergunta a idade de uma pessoa. Se a idade for maior do que 18 anos, o programa imprime na tela o texto ”Oi! Você é um adulto.”
Data: 24/03/2026
Autor: Isadora A. Paizani
Versão: 0.0.1
"""
```
